### Load train dataset

```python
import pandas as pd
from pycorenlp import StanfordCoreNLP
```

```python
# set up StanfordCoreNlp
nlp = StanfordCoreNLP('http://localhost:9000')
```

```python
train_data = pd.read_excel('..\\dataset\\train.xlsx')
train_data.columns
```

```python
train_data = train_data.filter(items=['ID','Statement', 'Subject', 'Label','Context'])
```

```python
train_data
```

```python
# Create a dictionary to store data for ease manipulation
data_dict =  dict()
```

```python
# Define keys and values
data_dict['ids'] = train_data[['ID']].values[:,0]
data_dict['labels'] = train_data[['Label']].values[:,0]
data_dict['statements'] = train_data[['Statement']].values[:,0]
data_dict['subjects'] = train_data[['Subject']].values[:,0]
data_dict['contexts'] = train_data[['Context']].values[:,0]
```

```python
# Process each statement
# Obtain the Part Of Speech for every word in the statement
# Count the number of words per statement
# Get the sentiment related to the statement (CoreNLP)
# Get POS bigram way
# Get POS trigram way

word_pos_list = list()
pos_list = list()
word_count_list = list()
sentiment_list = list()
sentiment_value_list = list()
bigram_pos_list = list()
trigram_pos_list = list()
for statement_id, statement in enumerate(data_dict['statements']):
    print('Procesing sentence number {0}: {1}'.format(statement_id,statement))
    output = nlp.annotate(statement, properties={
      'annotators': 'tokenize,pos,parse,sentiment',
      'outputFormat': 'json'
      })
    result_word_pos = str()
    result_pos = list()
    word_count = 0
    sentiment = str()
    sentiment_value = str()
    bigram_pos = list()
    trigram_pos = list()
    for o in output['sentences']:
        sentiment = o['sentiment']
        sentiment_value = o['sentimentValue']
        result_pos.append('<s>')
        for t in o['tokens']:
            word_count+=1
            result_word_pos += '{0}/{1} '.format(t['word'],t['pos'])
            result_pos.append('{0}'.format(t['pos']))
        for rpIndex, rp in enumerate(result_pos):
            if rpIndex < len(result_pos)-1 and len(result_pos)>=2:
                bigram_pos.append('{0} {1}'.format(rp, result_pos[rpIndex+1]))
            if rpIndex < len(result_pos)-2 and len(result_pos)>=3:
                trigram_pos.append('{0} {1} {2}'.format(rp, result_pos[rpIndex+1], result_pos[rpIndex+2]))
    word_pos_list.append(result_word_pos)
    pos_list.append(result_pos)
    word_count_list.append(word_count)
    sentiment_list.append(sentiment)
    sentiment_value_list.append(sentiment_value)
    bigram_pos_list.append(bigram_pos)
    trigram_pos_list.append(trigram_pos)
```

```python
data_dict['word_pos'] = word_pos_list
data_dict['pos'] = pos_list
data_dict['w_counts'] = word_count_list
data_dict['sentiments'] = sentiment_list
data_dict['sentiment_values'] = sentiment_value_list
data_dict['bigrams'] = bigram_pos_list
data_dict['trigrams'] = trigram_pos_list
```

```python
train_data['Word POS'] = data_dict['word_pos']
train_data['POS'] = data_dict['pos']
train_data['Word Count'] = data_dict['w_counts']
train_data['Sentiment'] = data_dict['sentiments']
train_data['Sentiment Value'] = data_dict['sentiment_values']
train_data['Bigram'] = data_dict['bigrams']
train_data['Trigram'] = data_dict['trigrams']
```

```python
pos_count = dict()
for sl in set(data_dict['labels']):
    lbl_value = str(sl).lower()
    pos_count[lbl_value] = dict()
    pos_count[lbl_value]['unigrams'] = dict()
    pos_count[lbl_value]['bigrams'] = dict()
    pos_count[lbl_value]['trigrams'] = dict()
```

```python
pos_count['false']['unigrams']
```

```python
for pos_index, uni_pos in enumerate(data_dict['pos']):
    pos_unigrams = uni_pos
    pos_bigrams = data_dict['bigrams'][pos_index]
    pos_trigrams = data_dict['trigrams'][pos_index]
    statement_label = str(data_dict['labels'][pos_index]).lower()
    
    for uni_key in pos_unigrams:
        if uni_key not in pos_count[statement_label]['unigrams']:
            pos_count[statement_label]['unigrams'][uni_key] = 1
        else:
            pos_count[statement_label]['unigrams'][uni_key] += 1
            
    for bi_key in pos_bigrams:
        if bi_key not in pos_count[statement_label]['bigrams']:
            pos_count[statement_label]['bigrams'][bi_key] = 1
        else:
            pos_count[statement_label]['bigrams'][bi_key] += 1
            
    for tri_key in pos_trigrams:
        if tri_key not in pos_count[statement_label]['trigrams']:
            pos_count[statement_label]['trigrams'][tri_key] = 1
        else:
            pos_count[statement_label]['trigrams'][tri_key] += 1
```
